# Ejercicio 1. 

En el caso de dos categorias, la regla de decision de Bayes el error condicional
 esta dada por la ecuacion (7)
 $$P(error|x)=\min [P (\omega_1 | x), P (\omega_2 | x)]$$ 
 
 Incluso si las densidades a posteriori son continuas, esta forma de la condicional de errores casi siempre conduce a un integrando discontinuo en el calculo del error total en la ecuacion (5) $$P(error)=\int P(error|x)p(x) dx$$

* (a) Demostrar que para densidades arbitrarias, podemos obtener una cota superior para el error total usando en la ecuacion anterior $$P(error|x)=2P (\omega_1 | x) P (\omega_2 | x)$$


**Respuesta**

Si suponemos que para un determinado $x$ se da que $P (\omega_1 | x) \geq P (\omega_2 | x)$ luego por la regla anterior se cumple que $P(error|x)= P (\omega_2 | x)$.

Como sabemos $P (\omega_1 | x) + P (\omega_2 | x) = 1$, entonces se da que $P (\omega_2 | x) = 1 - P (\omega_1 | x)$

Luego tenemos que $P (\omega_1 | x) \geq 1 - P (\omega_1 | x) \iff 2P (\omega_1 | x) \geq 1$. De lo que pobemos obtener:

$$2P (\omega_1 | x)P (\omega_2 | x) \geq P (\omega_2 | x) = P (error | x)$$

$$\iff$$

$$2P (\omega_1 | x)P (\omega_2 | x) p(x) \geq  P (error | x) p(x)$$

Si esto vale para todo $x$ entonces con las integrales también se cumple, por lo tanto:

$$\int 2P (\omega_1 | x)P (\omega_2 | x) p(x) dx \geq  \int P (error | x) p(x)dx = P(error) $$


* (b) Demostrar que si en la ecuacion (5), utilizamos $P (error | x) =\alpha P (\omega_1 | x) P (\omega_2 | x)$, con $\alpha<2$, entonces no podemos garantizar que la integral de una cota superior para el error.

**Respuesta**

Se busca un contraejemplo, si tomamos $$P (\omega_1 | x)= 0.7, P (\omega_2 | x)=0.3,  \alpha = 0.5 $$
Entonces:

$$ \alpha P (\omega_1 | x)P (\omega_2 | x) = 0.5 \times 0.7 \times 0.3 = 0.105 < P (error | x) = 0.3 $$

* (c) Analogamente, demostrar que podemos utilizar $$P (error | x) = P (\omega_1 | x) P (\omega_2 | x)$$ y obtener una cota inferior para el error total.

**Respuesta**

Si suponemos que para un determinado $x$ se da que $P (\omega_1 | x) \geq P (\omega_2 | x)$ luego por la regla anterior se cumple que $P(error|x)= P (\omega_2 | x)$.

Ademas sabemos que $$P(\omega_1|x) \leq 1$$

$$\iff$$

$$P(\omega_1|x)P(\omega_2|x) \leq P(\omega_2|x) = P(error|x)$$

$$\iff$$

$$P(\omega_1|x)P(\omega_2|x)p(x) \leq P(error|x)p(x)$$

Si esto vale para todo $x$ entonces con las integrales también se cumple, por lo tanto:

$$\int P (\omega_1 | x)P (\omega_2 | x) p(x) dx \leq  \int P (error | x) p(x)dx = P(error) $$


* (d) Demostrar que si $P (error | x) =\beta P (\omega_1 | x) P (\omega_2 | x)$ con $\beta> 1$, entonces no podemos garantizar que la integral de una cota inferior para el error.

**Respuesta**

Se busca un contraejemplo, si tomamos $$P (\omega_1 | x)= 0.6, P (\omega_2 | x)=0.4,  \beta = 2 $$
Entonces:

$$ \beta P (\omega_1 | x)P (\omega_2 | x) = 2 \times 0.6 \times 0.4 = 0.48 > P (error | x) = 0.4 $$


# Ejercicio 2. 

Suponga que se tienen dos variables independiente con igual funcion densidad
$$
p(x|\omega_i) = \propto e^{(-|x-a_i|/b_i)}
$$
para $i=1,2$ and $0<b_i$.

* (a)  Escriba una expresion analitica para cada densidad, es decir normalize cada funcion para parametros $a_i, b_i$ arbitrarios, $b_i$ positivo.

**Respuesta**

$$\int_{-\infty}^{\infty} p(x|\omega_i) dx = \int_{-\infty}^{\infty}  \propto e^{(-|x-a_i|/b_i)} dx$$

 $$ = {} \propto \left[ \int_{-\infty}^{\infty}  e^{(-|x-a_i|/b_i)} dx \right] = 1$$
 
 Busca el valor que de $\propto$ que noramalize la densidad.
 
 Por la definicion de valor absoluto: 
 
 $$ |x - a_i| = \left\{ \begin{array}{lcc}
             x - a_i &   si  & x \geq a_i \\
             a_i - x &  si &  x < a_i \\
             \end{array}
   \right.$$

La integral nos queda:

$$ \propto \left[ \int_{-\infty}^{\infty}  e^{(-|x-a_i|/b_i)} dx \right] = {} 
    \propto \left[ \int_{-\infty}^{a_i}  e^{(-(a_i-x)/b_i)} dx  + 
                   \int_{a_i}^{\infty}  e^{(-(x-a_i)/b_i)} dx 
            \right] = 1 $$

Resolviendo cada una de las integrales en la suma, obtenemos:

$$ \propto \left[ \int_{-\infty}^{\infty}  e^{(-|x-a_i|/b_i)} dx \right] = {} 
    \propto \left[ b_i + b_i
            \right] = 1 $$

$$\iff$$
 
$$\propto {} = \dfrac{1}{2b_i}$$

Por lo que la expresión para la densidad es : 

$$p(x|\omega_i) = \dfrac{1}{2b_i} e^{(-|x-a_i|/b_i)}
 $$

Se correponde con una [Distribución de Laplace](https://en.wikipedia.org/wiki/Laplace_distribution)

* (b) Calcule el radio de verosimilitud como función de los parametros.

$$\dfrac{p(x|\omega_1)}{p(x|\omega_2)} = \dfrac{b_2}{b_1}exp \left[ \dfrac{|x-a_2|}{b_2} - \dfrac{|x-a_1|}{b_1}\right]  $$

* (c) Graficar el radio $p(x|\omega_1)/p(x|\omega_2)$ para el caso $a_1=0$, $b_1=1$, $a_2=1$ y $b_2=2$.

**Respuesta**

$$\dfrac{p(x|\omega_1)}{p(x|\omega_2)} = 2exp \left[ \dfrac{|x-1|}{2} - |x|\right]  $$

Sacando los valores absoluto el radio de verosimilitud queda defindo como:

$$ \dfrac{p(x|\omega_1)}{p(x|\omega_2)} = \left\{ \begin{array}{lcc}
             2e^{(1+x)/2} &   si  & x < 0 \\
             2e^{(1-3x)/2} & si  & 0 \leq x < 1 \\
              2e^{(-1-x)/2} &  si &  x \geq 1 \\
             \end{array}
   \right.$$



```python
import matplotlib.pyplot as plt
import numpy as np

def likelihood_ratio_1(x):
    return 2*np.exp((1+x)/2)

def likelihood_ratio_2(x):
    return 2*np.exp((1-3*x)/2)

def likelihood_ratio_3(x):
    return 2*np.exp((-1-x)/2)

x_1 = np.linspace(-5, 0, 300, endpoint=False)
x_2 = np.linspace(0, 1, 300, endpoint=False)
x_3 = np.linspace(1, 5, 300)

plt.figure(figsize=(7,5))
plt.plot(x_1, likelihood_ratio_1(x_1))
plt.plot(x_2, likelihood_ratio_2(x_2))
plt.plot(x_3, likelihood_ratio_3(x_3))
plt.ylabel('Likelihood ratio');
```

# Ejercicio 3. 

Si las distribuciones condicionales para las dos categorias en el problema unidimensional son distribuciones de Cauchy
$$
p(x|\omega_i)=\frac{1}{\pi b}. \frac{1}{1+\left(\frac{x-a_i}{b}\right)^2}
$$


* (a) Suponiendo que $P(\omega_1) = P(\omega_2)$, muestre que $P(\omega_1 | x) = P (\omega_2 | x)$ si $x =\frac{a_1 + a_2}{2}$, es decir, el minimo para las cotas de la probabilidad de error en la decision es el punto medio entre los picos de las dos distribuciones, independientemente de la b.

**Respuesta**

$$P(\omega_1 | x) = P (\omega_2 | x) \iff \dfrac{p(x|\omega_1)P(\omega_1)}{p(x)} = \dfrac{p(x|\omega_2)P(\omega_2)}{p(x)} \iff p(\omega_1|x) = p(\omega_2|x)$$
<p style='text-align: center;'> (ya que las probabilidades a priori son iguales) </p>
<p style='text-align: right;'>  </p>

$$\iff$$

$$\frac{1}{\pi b}. \frac{1}{1+\left(\frac{x-a_1}{b}\right)^2} = \frac{1}{\pi b}. \frac{1}{1+\left(\frac{x-a_2}{b}\right)^2} \iff |x- a_1| = |x - a_2| $$


Si $a_1 \neq a_2$ entonces esto solo ocurre si $x = \dfrac{a_1 + a_2}{2}$


* (b)  Graficar $P(\omega_1|x)$ para el caso de $a_1=3$, $a_2=5$ y $b = 1$.

Por teorema de Bayes se tiene: 

$$P(\omega_1|x) = \dfrac{p(x|\omega_1)P(\omega_1)}{p(x)} = \dfrac{p(x|\omega_1)P(\omega_1)}{p(x|\omega_1)P(\omega_1) + p(x|\omega_2)P(\omega_2)}$$

```python
from scipy.stats import cauchy

def posteriori_omega(x, a_1, a_2, b):
    likelihood = cauchy.pdf(x, loc=a_1, scale=b)
    evidence = 0.5*(cauchy.pdf(x, loc=a_1, scale=1) + cauchy.pdf(x, loc=a_2, scale=b))  
    priori = 0.5 
    return (likelihood*priori)/evidence

x = np.linspace(-20, 30, 300)
plt.figure(figsize=(7,5))
plt.plot(x, posteriori_omega(x, 3, 5, 1))
plt.title('Posterior probabilty for omega_1')
plt.ylabel('probability');
```

Observamos que $P(\omega_1|x) = \dfrac{1 + (x-5)^{2}}{2 + (x-3)^{2} + (x-5)^{2}}$ con los valores de $a_1$, $a_2$ y $b$ del ejercicio anterior y las probabilidades a priori iguales. 

Por lo tanto:

$$\lim\limits_{x \to \infty}P(\omega_1|x) = \lim\limits_{x \to \infty}\dfrac{1 + (x-5)^{2}}{2 + (x-3)^{2} + (x-5)^{2}} = \dfrac{1}{2}$$

De la misma forma $\lim\limits_{x \to -\infty}P(\omega_1|x) = \dfrac{1}{2}$, lo que se confirma mirando el gráfico anterior.

Si las distribuciones condicionales para las dos categorias en el problema unidimensional son distribuciones de Cauchy, y asuma la igualdad entre las probabilidades a priori para las categorias.

* (a) Demostrar que la probabilidad de error minimo esta dado por $$P (error) =\displaystyle\frac12-\displaystyle\frac{1}{\pi}\tan^{-1}\left|\displaystyle\frac{a_2-a_1}{2b}\right|$$

**Respuesta**

Teniamos que: 
    $$P(error)=\int_{-\infty}^{\infty} P(error|x)p(x) dx$$
    
y que la regla de decisión de Bayes para el error condicional $P(error|x)=\min [P (\omega_1 | x), P (\omega_2 | x)]$ me aseguraba el mínimo error.

Por otro lado, en el ejercicio anterior inciso a) habíamos visto que cuando las probabilidades a priori eran iguales, el límite mínimo de decisión de error es un punto intermedio entre los picos de las dos distribuciones de Cauchy. 

Esto se observa en el siguiente grafico en donde se dibujan las dos probabilidades a *posteriori* para $\omega_1$ con $a_1 = 3$ y $\omega_2$ con $a_2 = 5$ con escala $b = 1$.

```python
x = np.linspace(-20, 30, 300)
a_1, a_2, b = 3, 5, 1
plt.figure(figsize=(11,6))
plt.plot(x, posteriori_omega(x, a_1, a_2, b),
         label='Posteriori omega_1')
plt.plot(x, posteriori_omega(x, a_2, a_1, b), 
         label='Posteriori omega_2')
boundary = (a_1 + a_2)/2
prob_boundary = posteriori_omega(boundary, a_1, a_2, b)
plt.plot(boundary, prob_boundary, 'ro', label='Boundary point')
plt.title('Posterior probabilties')
plt.ylabel('probability')
plt.legend();
```

A la izquierda de *Boundary point*, $P(\omega_1|x)$ está por encima de $P(\omega_2|x)$ y la derecha es al revés; entonces podemos definir cuanto vale $P(error|x)$ en cada caso. Luego, si $m=\dfrac{a_1 + a_2}{2}$:

 $$P(error)=\int_{-\infty}^{\infty} P(error|x)p(x) dx = 
            \int_{-\infty}^{m} P(\omega_2|x)p(x) dx + \int_{m}^{\infty} P(\omega_1|x)p(x) dx $$
 
 $$ = \int_{-\infty}^{m} p(x|\omega_2)P(\omega_2) dx + \int_{m}^{\infty} p(\omega_1|x)P(\omega_1) dx$$
 
 $$ = \dfrac{1}{2} \left[\int_{-\infty}^{m} p(x|\omega_2) dx + \int_{m}^{\infty} p(\omega_1|x) dx\right]$$
 
 $$ = \dfrac{1}{2} \frac{1}{\pi b} \left[\int_{-\infty}^{m} \frac{1}{1+\left(\dfrac{x-a_2}{b}\right)^2} dx + \int_{m}^{\infty}  \frac{1}{1+\left(\dfrac{x-a_1}{b}\right)^2} dx\right]$$
 
Realizando la sustucion $u = \dfrac{x-a_i}{b}$ en cada una de las integrales, se obtiene:
 
  $$ P(error) = \dfrac{1}{2} \frac{1}{\pi} \left[\tan^{-1}\left(\dfrac{a_1 - a_2}{2b}\right) + \dfrac{\pi}{2} + \dfrac{\pi}{2} - \tan^{-1}\left(\dfrac{a_2 - a_1}{2b}\right)\right]$$
  
  $$ = \dfrac{1}{2} \frac{1}{\pi} \left[\pi -\tan^{-1}\left|\dfrac{a_2 - a_1}{2b}\right| -\tan^{-1}\left|\dfrac{a_2 - a_1}{2b}\right|\right]$$
  
  $$ = \dfrac{1}{2} - \frac{1}{\pi}\tan^{-1}\left|\dfrac{a_2 - a_1}{2b}\right|$$

* (b)  Graficar esto como una funcion de $\displaystyle\frac{a_2-a_1}{2b}$.

```python
def prob_error(x):
    return 0.5-(1/np.pi)*np.arctan(abs(x))

x = np.linspace(-7, 7, 200)
plt.figure(figsize=(7,5))
plt.plot(x, prob_error(x))
plt.title('Probability of error in function of (a_2-a_1)/2b')
plt.ylabel('probability');
```

**Respuesta**

El máximo error posible es 0.5, lo que ourre cuando $\dfrac{a_2 - a_1}{2b} = 0$, es decir, cuando $a_2 = a_1$ o bien cuando $b \rightarrow \infty$
