### Ejercicio 1: Version Spaces and Conjunctive Hypotheses

### (a) Qué son los elementos de un espacio de versión? Cómo están ordenados? Qué se puede decir sobre la definición y el tamaño de  S and G?

**¿Qué son los elmentos de un espacio de versión?**

El **espacio de versión** está formado por **Hipótesis**. Recordemos que una hipotesis era un conjunto de restricciones sobre las variables medidas. Las restricciones posibles para cada hipótesis podían ser $\emptyset, \$ $ y los *valores* que toman los atributos estudiados. 

Sin embargo, el espacio de versión contiene solo alquellas hipotesis que son **consistentes** con un conjunto de ejemplos de entrenamiento $D$. Luego, que cumplen con la siguiente definición: 

* Una hipótesis $h$ es **consistente** con un conjunto de ejemplos de entrenamiento $D$ sí y solo sí $h(x) = c(x)$ para cada ejemplo $<x, c(x)>$ en $D$. Es decir, 

    $Consistent(h, D) = (\forall<x, c(x)> \in D) h(x) = c(x)$

Por lo tanto, **espacio de versión**, denotado como $VS_{H, D}$, con respecto a un espacio de hipótesis $H$ y a un conjunto de ejemplos de entrenamiento $D$, es el **subconjunto** de hipótesis de $H$ **consistentes** con el conjunto de ejemplos de entrenamiento $D$. Es decir, 
                              $VS_{H,D} = \{h \in H | Consistent(h, D) \}$

<br>

**¿Cómo están ordenados?**

Las hipotesis del espacio de versión pueden ordenarse siguiendo la relación $\geq_{g}$ que define un **orden parcial** (reflexiva, simétrica, transitiva) sobre el espacio de hipótesis $H$.

La definicion de $\geq_{g}$ es la siguiente:

* Dadas dos hipótesis $h_{j}$ y $h_{k}$, decimos que $h_{j}$ es *mas_general_que_o_igual_a* $h_{k}$ sí y sólo si cualquier instancia $x \in X$ (conjunto de instancias posibles) que satisface $h_{j}$ (ie, $h_{j}(x) = 1$) también la satisface $h_{k}$. 

    Formalmente: $h_{j}$ es *mas_general_que_o_igual_a* $h_{k}$ sí y sólo si $(\forall x \in X)[h_{j}(x) = 1 \rightarrow h_{k}(x) = 1]$


* A partir del concepto anterior también podemos tener la relacion *mas_general_que* entre dos hipótesis $h_{j}$ y $h_{k}$, escrita como $h_{j} >_{g} h_{k}$, sí y solo sí $(h_{j} \geq_{g} h_{k})$ y $(h_{k} \ngeq_{g} h_{j})$.


* También diremos que $h_{j}$ es *mas_especifica_que* $h_{k}$ sí y sólo sí $h_{k}$ es *mas_general_que* $h_{j}$.

<br>

**¿Qué se puede decir sobre la definición y el tamaño de S and G?**

$G$ y $S$ son los conjuntos que contienen las hipotesis de máxima generalidad y máxima específicidad del espacio de hipótesis $H$ consistentes con una muestra de entrenamiento $D$, respectivamente.

Formalmente: 

 * $G = \{ g \in H | Consistent(g, D)  \land  (\neg\exists g^{'} \in H)[(g^{'} >_{g} g) \land Consistent(g^{'}, D)]\}$
 
 * $S = \{ s \in H | Consistent(s, D)  \land  (\neg\exists s^{'} \in H)[(s >_{g} s^{'}) \land Consistent(s^{'}, D)]\}$

Si $G$ y $S$ están bien definidas, pueden especificar completamente el espacio de versión. El teorema de representacion  nos dice que el espacio de version es el conjunto de hipótesis contenido en $G$ y $S$ más aquellas hipótesis que están entre $S$ y $G$ en el espacio de hipótesis ordenadas parcialmente.

Sobre el tamaño de $G$ y $S$:

*  Si $|S| \neq 0$ y $|G| \neq 0$, resultan ser respectivamente la cota
especifica y cota general del espacio de versiones (respecto del
conjunto de entrenamiento $D$).

* Si $|S| = |G| = 1$ entonces se encontró una unica hipótesis $h$ consistente con todos los ejemplos de entrenamiento.

* Si $|S| = |G| = 0$, no existe $h \in H$ consistente con los ejemplos.
